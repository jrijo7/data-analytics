Crie um relatório de vendas formatado usando f-strings com os seguintes dados:

- Nome do produto: "Notebook Gamer"
- Quantidade vendida: 125
- Preço unitário: 3299.99
- Desconto aplicado: 0.15 (15%)
- Total bruto: quantidade × preço
- Total com desconto: total bruto × (1 - desconto)

Exiba um relatório formatado com:
1. Nome do produto centralizado (20 caracteres)
2. Quantidade com preenchimento de zeros (3 dígitos)
3. Preço unitário como moeda (R$ com 2 decimais)
4. Desconto como porcentagem (2 decimais)
5. Total bruto com separador de milhar e 2 decimais
6. Total com desconto com separador de milhar e 2 decimais

**Saída esperada**:
```
========== RELATÓRIO DE VENDAS ==========
Produto:    Notebook Gamer    
Quantidade: 125 unidades
```

**Dicas**:

Crie um sistema que valide se uma pessoa pode dirigir baseado nas seguintes condições:

- Idade mínima: 18 anos

- Não está com a carteira suspensa: True/False

A pessoa pode dirigir se:
- Tiver 18 anos ou mais **E**

- Não estiver com a carteira suspensa

### Exercício 3.4 - Calculadora de Desconto Progressivo

Crie uma calculadora de desconto que aplica descontos progressivos baseados no valor da compra:

- Compras até R$ 100: sem desconto
- Compras entre R$ 100 e R$ 500: 5% de desconto
- Compras entre R$ 500 e R$ 1000: 10% de desconto
- Compras acima de R$ 1000: 15% de desconto

Além disso, se o cliente for VIP (True/False), adicione 2% de desconto extra.

O programa deve:
1. Solicitar o valor da compra
2. Solicitar se é cliente VIP
3. Calcular o desconto baseado na faixa de valor
4. Adicionar desconto VIP se aplicável
5. Calcular o valor final
6. Exibir: valor original, desconto aplicado (%), valor do desconto (R$), valor final

**Exemplo de entrada**:
```
Digite o valor da compra: R$ 750.00
É cliente VIP? (True/False): True
```

**Saída esperada**:
```
Valor original: R$ 750.00
Desconto aplicado: 12.00%
Valor do desconto: R$ 90.00
Valor final: R$ 660.00
```

**Dicas**:
- Use operadores relacionais para verificar faixas (`>`, `<=`, `and`)
- Use operadores aritméticos para cálculos
- Combine múltiplas condições com operadores lógicos
- Lembre-se da precedência: parênteses > aritméticos > relacionais > lógicos

````python
valor_original = float(input("Digite o valor da compra: R$"))
vip = input("É cliente VIP? (True/False): ") == "True"

if valor_original <= 100:
  porcentagem_desconto = 0
elif valor_original > 100 and valor_original <=500:
  porcentagem_desconto = 5
elif valor_original > 500 and valor_original <= 1000:
  porcentagem_desconto = 10
else:
  porcentagem_desconto = 15

if vip:
  porcentagem_desconto += 2

valor_desconto = valor_original * (porcentagem_desconto / 100)
valor_final = valor_original - valor_desconto

print(f'Valor original: R$ {valor_original:.2f}')
print(f'Desconto aplicado: {porcentagem_desconto:.2f}%')
print(f'Valor do desconto: R$ {valor_desconto:.2f}')
print(f'Valor final: R$ {valor_final:.2f}')
````

### Exercício 4.3 - Validador de Senha com WHILE

Crie um programa que valide uma senha. A senha deve:

- Ter pelo menos 8 caracteres
- Conter pelo menos uma letra maiúscula
- Conter pelo menos uma letra minúscula
- Conter pelo menos um número

O programa deve solicitar a senha repetidamente até que uma senha válida seja fornecida, usando `while`.

**Exemplo de execução**:
```
Digite uma senha: abc
Senha inválida! Tente novamente.
Digite uma senha: senha123
Senha inválida! Tente novamente.
Digite uma senha: Senha123
Senha válida! Cadastro realizado com sucesso.
```

**Dicas**:
- Use `while` para repetir até condição ser satisfeita
- Use métodos de string como `isupper()`, `islower()`, `isnumeric()`
- Use `any()` com list comprehension ou loops para verificar condições
- Ou percorra a string com `for` para verificar cada caractere

````python
senha_valida = False

while not senha_valida:
  senha = input('Digite uma senha: ')

  tamanho_ok = len(senha) >= 8

  tem_maiuscula = False
  tem_minuscula = False
  tem_numero = False

  for caractere in senha:
    if caractere.isupper():
      tem_maiuscula = True
    elif caractere.islower():
      tem_minuscula = True
    elif caractere.isnumeric():
      tem_numero = True

  if tamanho_ok and tem_maiuscula and tem_minuscula and tem_numero:
    senha_valida = True
    print('Senha válida! Cadastro realizado com sucesso.')
  else:
    print('Senha inválida! Tente novamente.')
````

### Exercício 4.4 - Menu Interativo com While True

Crie um menu interativo que permite ao usuário:

1. Calcular área de um retângulo
2. Calcular área de um círculo
3. Verificar se um número é par ou ímpar
4. Sair do programa

O programa deve:
- Exibir um menu numerado
- Solicitar a opção do usuário
- Executar a operação escolhida
- Voltar ao menu após cada operação (exceto ao sair)
- Usar `while True` com `break` para sair
- Usar `continue` para voltar ao início do loop em caso de opção inválida

**Exemplo de execução**:
```
=== MENU ===
1. Calcular área do retângulo
2. Calcular área do círculo
3. Verificar se número é par/ímpar
4. Sair
Escolha uma opção: 1
Digite a largura: 5
Digite a altura: 3
Área do retângulo: 15.0

=== MENU ===
...
Escolha uma opção: 4
Programa encerrado. Até logo!
```

**Dicas**:
- Use `while True:` para loop infinito
- Use `if/elif/else` para tratar cada opção
- Use `break` para sair do loop
- Use `continue` para opções inválidas
- Para área do círculo: π × raio² (use 3.14159 para π)

````python
while True:
  print('=== MENU ===')
  print('1. Calcular área do retângulo')
  print('2. Calcular área do círculo')
  print('3. Verificar se número é par/ímpar')
  print('4. Sair')

  opcao = input('Escolha uma opção: ')

  if opcao == '1':
    largura = float(input('Digite a largura: '))
    altura = float(input('Digite a altura: '))
    area = largura * altura
    print(f'Área do retângulo: {area}')

  elif opcao == '2':
    raio = float(input('Digite o raio: '))
    area = 3.14159 * (raio ** 2)
    print(f'Área do círculo: {area}')

  elif opcao == '3':
    numero = int(input('Digite um número: '))
    if numero % 2 == 0:
      print(f'O número {numero} é par.')
    else:
      print(f'O número {numero} é ímpar.')

  elif opcao == '4':
    print('Programa encerrado. Até logo!')
    break

  else:
    print('Opção inválida. Tente novamente.')
    continue
````
